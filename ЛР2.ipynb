```python
# iloc - выделяет из таблицы указанный фрагмент, в данном случае все строки и только первый столбец
import pandas as pd     # для чтения и записи табличных файлов
import numpy as np      # для математической обработки данных
import matplotlib.pyplot as plt # для построения графиков
from sklearn.metrics import mean_squared_error
from math import sqrt
url = "https://drive.google.com/uc?export=download&id=1R-bDTTCsLjzy9NZG3bkzzeNnFyBuupGK"
# index_col показывает, какой столбец нужно использовать как индекс строки в таблице
input_df = pd.read_csv(url, sep = ';') 
product = input_df.iloc[: , 1].values
forecast = input_df.iloc[: , 0].values
k = 0.9
for i in range(2, len(product)):
  # делаем прогноз на i-й час, так как он еще как будто не наступил, то можем использовать наблюдения только за прошлые часы (i - 1, i - 2, ...) 
  forecast[i] = product[i - 1] * k + product[i - 2] * (1. - k)
input_df.dropna(axis=0, inplace=True)
mse = mean_squared_error(input_df.iloc[:,1], input_df.iloc[:,0])
rmse = np.sqrt(mse)
print(round(rmse,2))
fig, ax = plt.subplots(figsize = (20,20))

ax.plot(product, linewidth = 2, markersize = 5, marker = 'o', color = 'green')
ax.plot(forecast, linewidth = 2, markersize = 5, marker = 'o', color = 'blue')
ax.plot(np.abs(forecast - product), 'red')
ax.grid(True)
ax.set_xlabel('day') 
ax.set_ylabel('Мощь')

plt.show()
```
